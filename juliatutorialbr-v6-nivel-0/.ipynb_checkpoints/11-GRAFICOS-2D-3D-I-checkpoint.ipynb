# GRÁFICOS 2D-3D PARTE I

Existem diversos pacotes gráficos com Julia destinados aos mais variados tipos apresentação visual de dados. Os pacotes diferenciam entre si em função dos recursos de configuração, velocidade de processamento e facilidade de uso. Alguns pacotes são: [*PyPlot*](https://github.com/JuliaPy/PyPlot.jl), [*Plotly*](https://github.com/spencerlyon2/PlotlyJS.jl), [*GR*](https://github.com/jheinen/GR.jl),  [*GadFly*](http://gadflyjl.org/stable/) e muitos outros.

**PyPlot.jl**

<img src="Figuras/matplotlib.png" align="center" width="300">

*PyPlot* permite plotar gráficos 2D e 3D, e utiliza o pacote matplotlib do *Python* (via *PyCall*). Uma das vantagens do pacote *PyPlot* é a quantidade de opções de gráficos e configurações de exibição ([Link GitHub](https://github.com/JuliaPy/PyPlot.jl)). 

**GR.jl**

<img src="Figuras/gr.png" align="center" width="300">

Este módulo fornece uma interface de Julia para O *GR Framework*. *GR* é uma estrutura universal para aplicações de visualização multi-plataforma e foi desenvolvido pelo grupo *Scientific IT-Systems* no Instituto Peter Grünberg do centro de pesquisa Jülich. Os gráficos gerados pelo *GR* apesar de simples carregam rápido.

**PlotlyJS.jl**

<img src="Figuras/plotlyJS.png" align="center" width="300">

É uma interface para a biblioteca *Plotly.js*. É uma biblioteca gráfica de alto nível que permite criar gráficos interativos de alta qualidade, incluindo gráficos 3D, gráficos estatísticos e mapas SVG. É construído sobre as bibliotecatas *d3.js* e *stack.gl*. 


**GadFlay.jl**

<img src="Figuras/gadfly.png" align="center" width="150">

*Gadfly* é um sistema para plotar e visualizar gráficos totalmente escrito em *Julia*. É baseado na biblioteca *ggplot2* e nas ideias do livro *The Grammar of Graphics* de Leland Wilkinson. Foi idealizado por de Daniel C. Jones e agora é mantida pela comunidade. Apresenta gráficos muito bonitos e várias opções de vizualização, incluindo uma barra de zoom.

## PLOTS

Plots é um conjunto de ferramentas para construção e visualização de gráficos 2D e 3D. *Plots* trabalha de forma associado com um "backends", ou seja, um pacote gráfico (*PyPlot, GR, PlotlyJS* e outros) responsável por plotar o gráfico, de tal forma que *Plots* cria uma interface para este pacote . Uma grande vantagem do pacote `Plots` é que um mesmo conjunto de comandos pode ser utilizado para vários `Backends`, eliminando assim a necessidade e aprender vários comandos diferentes para cada pacote gráfico.

Alguns dos *backends* tem funcionalidades que outras não tem. O melhor lugar para aprender é na [documentação (em inglês)](https://docs.juliaplots.org/latest/), onde há uma série de [exemplos](https://docs.juliaplots.org/latest/examples/gr/).

```julia
using Plots
pacote_grafico() # pyplot, plotly, gr, UnicodePlots e PGFPlots
   tipo_grafico(parametros)
```

## GRÁFICOS 2D

### GRÁFICO DE DADOS DISCRETO (SCATTER)


Sintaxe:
```julia
using Plots 
pacote_grafico()

scatter(x_dados, y_dados, title = "nome_gráfico", label = "nome_pontos", opções)

scatter(matriz, title = "nome_gráfico", label = "nome_pontos", opções)

```

````julia
x1 = [0.0, 1.0, 1.5, 2.5, 3.0, 4.0, 4.5, 5.0, 6.0, 7.0, 7.5, 8.0];
y1 = [0.0, 0.84, 1.0, 0.6, 0.14, -0.76, -0.98, -0.96, -0.28, 0.66, 0.94, 0.99];
````

````julia
matriz_dados = [1.0  1.5  2.0; 1.0  2.0  3.0; 1.0  2.0  4.0]
````

````julia
using Plots
gr()

scatter(x1 , y1 )
````

````julia
scatter(matriz_dados)
````

**Melhorando o Gráfico**

````julia
using Plots
gr()

scatter(x1 , y1, label = "dados", size = (400, 300), xlabel = "x", ylabel = "y" )
````

Outro recurso interessante é a quantidade de pontos por polegada, "dpi". Quanto maior a "dpi" melhor a qualidade gráfica do gráfico. O padrão é 100.

````julia
scatter(x1 , y1, label = "dados", size = (400, 300), xlabel = "x", ylabel = "y", dpi = 70)
````

### GRÁFICO BARRA DE ERRO (Error Bar)

Uma barra de erro é uma linha indicadora que se estende sobre um intervalo de y - ValorNegativoDoErro até y + ValorPositivoDoErro. Neste termo, y é o valor do ponto de dados.

````julia
using Plots
gr()
````

````julia
x1 = [0.0, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0];
y1 = [0.0, 0.84, 1.0, 0.91, 0.6, 0.14, -0.35, -0.76, -0.98, -0.96, -0.71, -0.28, 0.22, 0.66, 0.94, 0.99];
````

````julia
# erro dos dados de y
erroY = 0.1 

scatter(x1, y1, yerr = erroY, size = (400, 300))
````

````julia
# erro dos dados de y1
erroY = 0.1 

# erro dos dados de x1
erroX = 0.5

scatter(x1, y1, yerr = erroY, xerr = erroX, size = (400, 300))
````

````julia
# erro dos dados de y1
erroY = [0.0,0.672,0.8,0.728,0.48,0.112,-0.28,-0.608,-0.784,-0.768,-0.568,-0.224,0.176,0.528,0.752,0.792]

scatter(x1, y1, yerr = erroY, size = (400, 300))
````

### GRÁFICOS MULTIPLOS 

#### GRÁFICOS GERADOS AO MESMO TEMPO SOBRE A MESMA FIGURA 

Usando-se `plot!`, adiciona-se um gráfico sobre (em conjunto) o último criado.

```julia
plot(x1_vetor, y1_vetor,label = "nome", outras_opções )
plot!(x2_vetor, y2_vetor,label = "nome", outras_opções  )

plot(var_vetor, função_f,label = "nome", outras_opções )
plot!(var_vetor, função_g,label = "nome", outras_opções )

plot(matriz, title = "nome_gráfico", label = "nome", outras_opções) 
```

````julia
vetor_x = 0 : 0.5 : 3*pi
fg1(x) = sin(x)*x
fg2(x) = cos(x)*x
````

````julia
using Plots
gr()
````

````julia
plot( vetor_x , fg1, label = "sin(x)*x")
plot!(vetor_x , fg2, label = "cos(x)*x", size = (500, 300), 
    xlabel="x" , ylabel="y") 
````

**Outra maneira**
```julia
plot([funcao1, funcao2], vetor_x, label = ["nome" "nome"])
```

````julia
using Plots
gr()
````

````julia
plot([fg1 , fg2], vetor_x, label = ["sin(x)*x" "cos(x)*x"], 
    size = (500, 300), 
    xlabel="x" , 
    ylabel="y")
````

### GRÁFICOS MÚLTIPLOS LADO A LADO


````julia
# funções
fg4(x) = sin(x)*x
fg5(x) = cos(x)*x
````

````julia
using Plots
gr()

vetor_x = 0 : 0.5 : 3*pi

p1 = plot(vetor_x, fg4, title = "sin(x)*x", label = "sin(x)*x")
p2 = plot(vetor_x, fg5, title = "cos(x)*x", label = "cos(x)*x", line = (:red))

plot(p1, p2, size = (600, 300))
````

## GRÁFICOS 3D


### GRÁFICO DE LINHAS

Sintaxe:
```julia
plot( x_dados , y_dados , z_dados, opções)    
```
Sendo:
* **x_dados**: vetor ou função do eixo x;
* **y_dados**: vetor ou função do eixo y;
* **z_dados**: vetor ou função de eixo z;
* **opções**;

````julia
using Plots
pyplot()
````

````julia
funcao_y(x) = x^2
funcao_z(x) = x^3
````

````julia
vetor_x = 1:1:10;
````

````julia
plot(vetor_x , funcao_y.(vetor_x), funcao_z.(vetor_x), label = "Linha", size = (400, 300) )
````

````julia
using Plots
pyplot()
````

````julia
z(x, y) = cos(x) - sin(y)
````

````julia
vetor_x = 0:0.5:10
vetor_y = 0:0.5:10;
````

````julia
plot(vetor_x, vetor_y, z.(vetor_x, vetor_y), label = "Linha", size = (400, 300) )
````

### GRÁFICO DE SUPERFÍCIE

Por padrão, a figura será colorida em tons de uma cor sólida, mas também suporta mapeamento de cores.

Sintaxe:
```julia
plot( x_dados , y_dados , z_dados, opções)    
```
Sendo:
* **x_dados**: vetor ou função do eixo x;
* **y_dados**: vetor ou função do eixo y;
* **z_dados**: vetor ou função de eixo z;
* **opções**;

````julia
using Plots
pyplot()
````

````julia
fz(x, y) = cos(x) - sin(y)
````

````julia
vetor_x = 0:0.5:10
vetor_y = 0:0.5:10
````

````julia
surface(vetor_x, vetor_y, fz, size = (400, 300), 
    xlabel = "x", ylabel = "y", zlabel = "z")
````

**Melhorando o gráfico 3D **


````julia
surface(vetor_x, vetor_y, fz, size = (400, 300), 
xlabel = "x", ylabel = "y", zlabel = "z")
````
