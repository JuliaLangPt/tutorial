## MÁQUINA LOCAL

**1. LINUX**

* **VIA GERENCIADOR DE SOFTWARES**

Digite Julia no gerenciador de software (Apper, Synaptic, ...) e veja se mostra a versão 1.0.0 ou superior. Se não aparecer, instale via terminal a partir do repositório da distribuição. Para compilar Julia são necessários que os seguinte software instalados no computador local:

- GNU make
- Gcc & g ++ (> = 4.7) ou Clang (> = 3.1, Xcode 4.3.3 no OS X) 
- Python (> = 2.7) - necessário para construir LLVM.
- Gfortran - compilação e vinculação de bibliotecas Fortran
- Perl - pré-processamento de arquivos.
- M4 - necessário para construir GMP.
- Cmake - necessário para criar o libgit2.



* **VIA TERMINAL**

**Ubuntu e Derivados (Lubuntu, Kubuntu, Mint, Zorin, Ementary, Linux Lite e outros)**

Antes de instalar é necessário as ferramentes de compilação 

```sh
 sudo apt-get update 
 sudo apt-get install build-essential
```

Instalação Julia: 
```sh
 sudo add-apt-repository ppa:staticfloat/juliareleases
 sudo add-apt-repository ppa:staticfloat/julia-deps
 sudo apt-get update 
 sudo apt-get install julia
```
     
**Debian e Derivados (Deepin, NetRunner, Maui Linux, SteamOS, Metamorphose Linux[Br] e outros )**

Antes de instalar é necessário as ferramentes de compilação:
```sh
 sudo apt-get update 
 sudo apt-get install build-essential
```
Detalhe importante: O Debian poderá manter em seus repositórios uma versão anterior para priorizar a estabilidade do sistema.
```sh
 sudo apt-get install julia
```
        
**RedHat e Derivados (CentOS, Scientific Linux, Fedora e outros)**

As distro Scientific Linux  e CentoOS utilizam o yum como gerenciador padrão de pacotes. Antes de instalar é necessário as ferramentes de compilação:
```sh
 sudo dnf/yum grouplist 
```
Veja o nome correspondente a "Development Tools and Libraries" ou "Ferramentas de Desenvolvimento C e Bibliotecas".
```sh
 sudo dnf/yum groupinstall "Development Tools and Libraries" ou 
 "Ferramentas de Desenvolvimento C e Bibliotecas"
```
Instalação Julia: 
```sh
 sudo dnf/yum copr enable nalimilan/julia
 sudo dnf/yum install julia
```        

**Arch e Derivados (Manjaro, Antergos, Chakra e outros)**

Antes de instalar é necessário as ferramentes de compilação:
```sh
 sudo pacman -Sy base-devel
 sudo pacman -Syu
```

Instalação Julia:
```sh
 sudo pacman -S Julia
```

**OpenSuse (como Root)**

Antes de instalar é necessário as ferramentes de compilação:
```sh
 zypper install -t pattern devel_basis
```

Instalação Julia: 
```sh
zypper addrepo http://download.opensuse.org/repositories/science/openSUSE_AQUI_SUA_VERSÃO/science.repo
zypper refresh
zypper install julia
```
**Slackware e Derivados (Salix, Absolute Linux, Porteus e outros)**

A instalação no Slackware segue o modo de instalação via binários. 

**Instalando via GIT (Administrador ou super usuário)**
```sh
 git clone git://github.com/JuliaLang/julia.git
 cd julia
 make
 ./install/bin/julia    
```  

Ao executar o comando "make" este baixará vários outros arquivos, portanto é necessário conexão com internet. Se ocorrer:

>  Failed to execute git ls-remote --tags --heads git://github.com/twbs/bootstrap-sass.git , exit code of #128 fatal: unable to connect to github.com: github.com[0: 192.30.252.130]: errno=Connection timed out

Use o comando abaixo e execute os comandos de instalação novamente:

```sh
 git config --global url."https://".insteadOf git://
```

se ocorrer o erro:

> WARNING: ZMQ had build errors.
> - packages with build errors remain installed in /home/jmp/.julia/v0.4 
> - build the package(s) and all dependencies with Pkg.build("ZMQ")
> - build a single package by running its deps/build.jl script 
 
significa que ha dependencia de algum recurso de compilação. Instale o "CodeLite" via Synaptics ou Apper que resolve. Ou instale as ferramentas de compilação visto acima.


**Usando Binários **

Você pode usar a linguagem Julia a partir dos binários, sem a necessidade de instalá-lo em sua máquina. Isso é útil pois basta baixar o arquivo `julia-X.X.X-linux-x86_64.tar.gz` do site, extrair para um diretório (em geral com o nome Julia) e executar o binário Julia:

```sh
 ./bin/julia
```

Caso ocorra erro, será necessário alterar o binário para executável como root ou super usurário:
```sh
 sudo chmod +x julia
```
Adicione um `alias` no bash. Execute no terminal:
```sh 
echo "alias julia='/home/usuario/pasta-julia/bin/julia'" >> ~/.bashrc && source ~/.bashrc
```
** Shell Script**

Uma forma prática de instalação é através do script escrito pelo professor da UFPR Abel Siqueira (http://abelsiqueira.github.io/). Basta copiar a linha de comando abaixo, colar no terminal e teclar enter:

```sh
bash -ci "$(curl -fsSL https://raw.githubusercontent.com/abelsiqueira/jill/master/jill.sh)"

```
O script fará todo o trabalho de baixar e compilar os binários.

## INTERFACES

### REPL

A linguagem Julia vem com um interpretador por linha de comando interativa REPL (read-eval-print loop) integrado ao executável julia. O REPL é um ótimo ambiente de programação para experimentar a linguagem mas não é o melhor ambiente para fazer trabalhos de programação - para isto, um editor de texto, ou ambiente de notebook interativo (por exemplo Jupyter) é uma escolha melhor. Apesar dos limites há vantagens em usar o REPL: é simples, extremamente leve, configurável e funciona sem qualquer instalação ou configuração extra.

Depois de instalar ou baixar o binário, execute o comando abaixo no terminal e aparecerá:

```sh
 julia  
```

Ou na pasta do binário

```sh
 ./julia
```

### IDE Gráficas

Existem várias interfaces gráficas para usar com Julia. Vejamos algumas:
* **JUNO** http://junolab.org/

`Juno` é um ambiente de desenvolvimento integrado (IDE) para a linguagem Julia. Ele fornece ferramentas poderosas para ajudá-lo a desenvolver código através de uma rica interface. Juno é construído no Atom, um excelente editor de texto disponível em .....
<img src="Figuras/juno-lab.png" align="center" width="700">

* **LiClipse** http://www.liclipse.com/

O `LiClipse` é um conjunto de plugins para aprimorar o `Eclipse`. `LiClipse` inclui temas de IDE aprimorados, editores embutidos para muitas linguagens, melhorias de usabilidade para todos os editores do Eclipse, empacotamento de alguns plugins comuns e instaladores que são nativamente integrados em cada plataforma.

<img src="Figuras/liclipse-julia.png" align="center" width="700">



* **SublimeText** https://www.sublimetext.com/3

É um editor de código-fonte proprietário multiplataforma escrito em `C++` e `Python`. Ele suporta nativamente muitas linguagens de programação e linguagens de marcação, e sua funcionalidade pode ser estendida por usuários com plugins, normalmente construídos pela comunidade e mantidos sob licenças de software livre.

<img src="Figuras/sublime-julia.png" align="center" width="700">


* **Jupyter** http://www.jupyter.org 

O `Jupyter` é uma aplicação web derivado do `IPython` que é um interpretador interativo para várias linguagens de programação, masespecialmente focado em `Python`. Ipython oferece "type introspection", "rich media", sintax shell, completação por tab e edição auxiliada por histórico de comando. `Jupyter` possui suporte para mais de 40 linguagens de programação, incluindo as mais populares em computação científica como `Python`, `R`, `Julia` e `Scala`. 

<img src="Figuras/jupyter-julia.png" align="center" width="700">


**Como Instalar o Jupyter**

A maneira mais fácil de instalar o `Jupyter` é através do `Anaconda` [https://www.continuum.io/downloads](https://www.continuum.io/downloads). Baixe a versão `Python 3.5.0` ou superior.

* **Instalação Windows**:

Faça o download, execute o arquivo e pronto. No terminal execute:
```cmd
C:\Users\Usuario>conda install jupyter
```

* **Instalação Linux**:

Após download, execute no terminal:

```sh
 sudo chmod +x Anaconda3-X.X.X-Linux-x86_64.sh    
 ./Anaconda3-X.X.X-Linux-x86_64.sh  
```

concorde com os termos e será iniciada a instalação. Após a instalação execute:

```sh
 conda intall jupyter
```

Uma máquina virtual (Virtual Machine – VM) pode ser definida como “uma duplicata eficiente e isolada de uma máquina real”. A IBM define uma máquina virtual como uma cópia isolada de um sistema físico, e essa cópia está totalmente 
protegida (LAUREANO, 2006). 

O software `VirtualBox` é uma aplicação de virtualização multi-plataforma que estende as capacidades de um computador para que este possa executar vários sistemas operacionais ao mesmo tempo. Você pode instalar várias máquinas virtuais de acordo com os limites físicos de memória RAM e disco rígido da máquina local. O `VitualBox` está disponível para várias plataformas e pode ser baixado no site www.virtualbox.org/wiki/Downloads . ## DOCKER

`Docker` é tecnologia alternativa de virtualização criada pela empresa `Docker,Inc` que permite criar ambientes isolados chamados de `containers`. "De forma bem resumida, podemos dizer que o `Docker` é uma plataforma aberta, criada com o objetivo de facilitar o desenvolvimento, a implantação e a execução de aplicações em ambientes isolados" (GOMES, 2017).

Para baixar uma imagem docker da linguagem Julia use: 

```sh
sudo docker pull jupyter/datascience-notebook
Status: Downloaded newer image for jupyter/datascience-notebook:latest
```
Após o download, execute:
```sh
sudo docker run -p 8888:8888 jupyter/datascience-notebook
           porta local : porta container
```
Inicie o navegador e utilize o endereço `localhost:8888`. Aparecerá a tela inicial do `jupyter`. Copie e colo e token no campo indicado no navegador e pronto para uso.
