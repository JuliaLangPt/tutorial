## TELEGRAM

<img src="Figuras/telegram-logo.png" align="center" width="100">


O Telegram é um aplicativo gratuito de mensagens instantânea com foco em velocidade e segurança, concorrente direto do WathsApp. O Telegram está disponível para smartphones ou tablets (Android, iOS, Windows Phone, Ubuntu Touch, Firefox OS), computadores (Windows, OSX, Linux, BSD e outros) e também como Aplicação web. Destaques importantes:

- Conversas criptografadas;
- Sincronização na nuvem entre dispositivos;
- Maior velocidade no envio de mensagens;
- Distribuição de servidores em todo o mundo para maior segurança e velocidade;
- API em código aberto e protocolo livre para todos;
- Gratuito, sem propaganda e mensalidades;
- Possui versões móveis, versões para desktop e navegador web;
- Pode enviar diversos tipos de arquivos de até 1.5Gb de tamanho;
- Canais públicos ou privados com número ilimitado de seguidores;
- Grupos com capacidade de até 5000 membros (super grupo)
- Mensagens de auto-destruição em chats privativos;
- Interação com bots (softwares robos);
- Possibilidade de adicionar um contato apenas com o nome de usuário.

### INSTALAR TELEGRAM-CLI VIA SNAPPY
Para instalar o telegram-cli, a maneira mais prática é através do Snappy. Snappy, de modo genérico, é um  um sistema de gerenciamento de pacotes que visa facilitar a instalação de pacotes e suas dependências.

**Fedora**:

```sh
sudo dnf install snapd
sudo snap install telegram-cli
```

**Ubuntu/Debian**:

```sh
sudo apt install snapd
sudo snap install telegram-cli
```

**Arch**:

```sh
yaourt -S snapd
```
Uma vez que o pacote snapd é instalado, você precisa ativar a unidade systemd:

```sh
sudo systemctl enable --now snapd.socket
sudo snap install telegram-cli
```
**OpenSuse**:

* OpenSuse 42.3:

```sh
sudo zypper addrepo --refresh http://download.opensuse.org/repositories/system:/snappy/openSUSE_Leap_42.3/snappy
```
* Tumbleweed:

```sh
sudo zypper addrepo --refresh http://download.opensuse.org/repositories/system:/snappy/openSUSE_Tumbleweed/snappy

```

* instalação Snap:
```sh
sudo zypper install snapd
```

Uma vez que o pacote snapd é instalado, você precisa ativar a unidade systemd:

```sh
sudo systemctl enable --now snapd.socket
sudo snap install telegram-cli

```


**Sintaxe**:
```julia
telegram-cli -k tg-server.pub -opções Tipo_de_mensagem
```

Opções:
- c : nome do arquivo de configuração
- d : modo daemon
- D : desabilitar a saída
- e : executa o comando e sai
- f : durante a autorização busca todas as mensagens desde o registro
- L : nome do arquivo de log
- N : mostrar o número das mensagens
- p : usar o perfil especificado
- R : desabilitar a leitura do teclado do usuário
- u : especificar username (poderá ser solicitado durante a autorização)
- v : nível de detalhamento -verbosity- (0-ERROR 1-WARNIN 2-NOTICE 3-DEBUG)
- W : mostra a fila do dialog_list e espera por resposta antes de tratar a entrada do teclado

Tipo de mensagem:

* Mensagem simples:  "msg nome_contato/grupo  mensagem"
* Fotos: "send_photo  nome_contato/grupo  nome_arquivo_foto" 
* Vídeo: "send_video  nome_contato/grupo  nome_arquivo_video"
* Texto: "send_document nome_contato/grupo  nome_arquivo"

Para iniciar o telegram-cli é necessário mandar uma mensagem pelo terminal. Será pedido o número do telefone e será enviado um código para o telefone contendo um código de login para ser inserido no terminal. Pronto, agora já pode utilizar no Jupyter.

## KDE CONNECT

O KDE Connect, de acordo com o site oficial, é um projeto para se comunicar em todos os dispositivos (Computador, celular, tablet e outros). Com o KDE Connect, é possível receber as notificações do celular no computador (ou do computador para o celular), controlar a reprodução de música no computador ou utilizar o celular como controle remoto. 

Passos:
1. Instalar o KDE Connect (Já deve estar instalado no KDE) no Linux e Android/IOS (via play.google.com ou Itunes)

```sh
sudo add-apt-repository ppa:vikoadi/ppa -y
sudo apt-get update
sudo apt-get install indicator-kdeconnect kdeconnect -y
```

Se o firewall instalado é o `ufw`, pode ser necessário abrir algumas portas:

```sh
sudo ufw allow 1714:1764/udp
sudo ufw allow 1714:1764/tcp
sudo ufw reload
```

2. Sincronizar

3. Opções:
    - Enviar Arquivos
    - Controle Multimedia
    - Introdução de dados remota
